# Классификация постов о спортивных событиях

Все посты относятся к двум категориям:
- hockey — хоккей;
- football — футбол.

В рамках задания вам предстоит обучить модель логистической регрессии для задачи классификации текстов. 

Модель должна различать посты о футболе и хоккее с использованием методов векторизации, изученных в предыдущих уроках.

```python
import pandas as pd
```

```python
data_path = 'https://code.s3.yandex.net/datasets/ds_s16_t5_dataset.csv'
```

```python
df = pd.read_csv(data_path)
print("Пост 1: ", df["text"][0])
print("\nПост 2: ", df["text"][1])
print("\nПост 3: ", df["text"][2])
```

```python
import pandas as pd
import nltk # импортируем библиотеку nltk для обработки текста
nltk.download('stopwords') # скачиваем все существующие стоп-слова

from nltk.corpus import stopwords # импортируем стоп-слова


russian_stopwords = set(stopwords.words('russian')) # Объявите здесь множество из стоп-слов русского языка

# Объявляем функцию для удаления стоп-слов
def remove_stopwords(text_tokens):
    filtered_sentence = [word for word in text_tokens if word not in russian_stopwords] # Оставьте здесь только те слова, которых нет во множестве russian_stopwords 
    return " ".join(filtered_sentence)


# df = pd.read_csv("ds_s16_t5_dataset.csv")

df['text'] = df['text'].str.replace(r'[^\w]|[\s+]', ' ', regex=True)  # Здесь замените все спец символы и лишние пробелы на один пробел
df['text'] = df['text'].str.strip()  #  Здесь удалите все переносы строк
df['text'] = df['text'].str.lower()  #  Здесь переведите тексты в нижний регистр

df['tokens'] = df['text'].str.split() # Здесь разделите тексты на токены по пробелам

df['filtered_sentences'] = df['tokens'].apply(remove_stopwords)  # Применяем ранее созданную функцию для удаления стоп-слов 

print(df['filtered_sentences'][0])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer


y = df["category"]  # Выделяем колонку с классами для предсказаний
x = df["filtered_sentences"]  #  Выделяем тексты с обработанными текстами

# Разделяем на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42) # Здесь разделите x, y на тренировочную и тестовую выборки

v = TfidfVectorizer()  # Здесь объявите векторизатор
v.fit_transform(X_train) # Здесь обучите векторизатор на данных x

X_train_vectorized = v.transform(X_train)  # Применяем векторизатор к обучающей выборке
X_test_vectorized = v.transform(X_test) # Здесь примените векторизатор к тестовой выборке
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report


clf = LogisticRegression(random_state=42, solver="sag")  # Здесь объявите логистическую регрессию с необходимыми параметрами
clf.fit(X_train_vectorized, y_train) # Здесь обучите модель на тренировочной выборке X_train_vectorized 

y_pred = clf.predict(X_test_vectorized) # здесь получите предсказания на тестовой выборке X_test_vectorized 

# Строим отчет по всем метрикам
report = classification_report(y_test, y_pred) # Здесь постройте отчет (y_test, y_pred) по метрикам
print(report)
```
