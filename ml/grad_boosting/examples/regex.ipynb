## Удаление стоп-слов

```python
import pandas as pd
import nltk # Импортируем библиотеку NLTK для обработки текста
nltk.download('stopwords') # Скачиваем все существующие стоп-слова

from nltk.corpus import stopwords # Импортируем стоп-слова
```

```python
russian_stopwords = set(stopwords.words("russian")) # Указываем, что нужно использовать только стоп-слова на русском языке

df = pd.read_csv("https://code.s3.yandex.net/datasets/ds_s16_t5_kinopoisk_reviews.csv")

random_positive_review = "594-16" # Определяем рецензию
review = df[df["id"] == random_positive_review] # Выбираем рецензию из датафрейма
print("Сырой текст: \n----------------------------------------------------------------------------------------------------------")
print(review["text"].to_list()[0]) # Получаем рецензию и печатаем
print("----------------------------------------------------------------------------------------------------------")
review_tokens = review["text"].to_list()[0].split() # Делим текст на токены
print(f"Рецензия длиной {len(review_tokens)} токена: {review_tokens}")
print("----------------------------------------------------------------------------------------------------------")
# Удаление стоп-слов
filtered_tokens = [
      word.lower() for word in review_tokens
      if word.lower() not in russian_stopwords
  ] # Фильтруем все стоп-слова
print(f"Рецензия без стоп-слов длиной {len(filtered_tokens)} токена: {filtered_tokens}")
```

## Bag of words 

Bag of Words — это метод векторизации текстов. Он преобразует текст в числовой вектор на основе встречаемости слов, игнорируя их порядок и грамматическую структуру. В алгоритме может использоваться как частота слов, так и просто их наличие в тексте.

```python
import re


text = """
Иван   любит смотреть   фильмы. Мария    тоже любит фильмы. Иван любит смотреть   футбол. Мария любит смотреть футбол. Иван и Мария друзья
""" # Определяем текст

dataset = text.split(".") # Делим текст по точкам

for i in range(len(dataset)):
    # Шаг 1. Приведение к нижнему регистру
    dataset[i] = dataset[i].lower()
    # Шаг 2. Удаление всех символов, кроме букв
    dataset[i] = re.sub(r'\W', ' ', dataset[i])
    # Шаг 3. Удаление лишних пробелов
    dataset[i] = re.sub(r'\s+', ' ', dataset[i])

for i, sentence in enumerate(dataset):
    print(f"Предложение {i+1}: {sentence}")
```

```python
import nltk
nltk.download('stopwords') # Скачиваем все стоп-слова
from nltk.corpus import stopwords
import pandas as pd

word2count = {} # Объявляем словарь

# Шаг 1. Формируем словарь с частотами слов
for data in dataset:
    words = data.split() # Токенизируем предложения по пробелам
    for word in words:
        if word not in word2count: # Считаем количество встречающихся слов
            word2count[word] = 1
        else:
            word2count[word] += 1

# Шаг 2. Объявляем стоп-слова на русском языке
stop_words = set(stopwords.words('russian'))

# Шаг 3. Фильтруем все слова по стоп-словам
filtered_word2count = {word: count for word, count in word2count.items() if word not in stop_words}

word_freq_df = pd.DataFrame(list(filtered_word2count.items()), columns=['Слово', 'Частота']) # Собираем датафрейм со словами и их частотами

word_freq_df = word_freq_df.sort_values(by='Частота', ascending=False) # Сортируем строки по частоте использования слов

print(word_freq_df)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

X = [] # Объявляем массив X

words = word_freq_df["Слово"].to_list()  # Формируем список из слов в словаре

for data in dataset:
    vector = []  # Объявляем пустой вектор
    for word in words:
        if word in data.split():  # Проверяем, есть ли слово в предложении
            vector.append(1)  # Слово есть — 1
        else:
            vector.append(0)  # Слова нет — 0
    X.append(vector)  # Добавляем вектор в массив X

X = np.asarray(X)

plt.figure(figsize=(10, 6))
sns.heatmap(X, cmap='RdYlGn', cbar=False, annot=True, fmt="d", xticklabels=words, yticklabels=[f"Предложение {i+1}" for i in range(len(dataset))])  # Объявляем график для отображения векторных представлений предложений

plt.title('Bag of Words матрица')
plt.xlabel('Слова в словаре')
plt.ylabel('Предложения')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import re


text = """
Иван   любит смотреть   фильмы. Мария    тоже любит фильмы. Иван любит смотреть   футбол. Мария любит смотреть футбол. Иван и Мария друзья
""" # Определяем текст

documents = text.split(".")  # Делим текст по точкам

for i in range(len(documents)):
    documents[i] = documents[i].lower() # Приводим весь текст к нижнему регистру
    documents[i] = re.sub(r'\W', ' ', documents[i]) # Оставляем только буквы
    documents[i] = re.sub(r'\s+', ' ', documents[i]) # Убираем лишние пробелы

# Создаём векторизатор
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents) # Обучаем векторизатор

# Определяем словарь
print("Словарь:", vectorizer.get_feature_names_out())

# Определяем векторы документов
print("Векторы:\n", X.toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import re


text = "В парке, на лавочке, лежала книга, то была забыта кем-то ещё утром. Я подумал, что это, а может, и не книга вовсе, а просто блокнот с записями. Но, открыв её, понял: то, что я нашёл, оказалось настоящим сокровищем — дневником путешественника, полным ярких впечатлений."

stop_words = ["а", "в", "на", "то", "что"]
documents = text.split('.')  # Здесь разделите текст на предложения по точкам

for i in range(len(documents)):
    documents[i] = documents[i].lower()  # Здесь приведите текст к нижнему регистру
    documents[i] = re.sub(r'\W', ' ', documents[i])  # Здесь оставьте только буквы

# Создаём векторизатор
vectorizer = CountVectorizer(stop_words=stop_words) #  Здесь объявите векторизатор со стоп-словами
X = vectorizer.fit_transform(documents) # Здесь примените векторизатор на текстах documents

# Определяем словарь
print("Словарь:", vectorizer.get_feature_names_out()) #  Здесь получите словарь

# Определяем векторы документов
print("Векторы:\n", X.toarray())
```

## Алгоритм векторизации: TF–IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Пример текстовых данных
text = """
Иван   любит смотреть   фильмы. Мария    тоже любит фильмы. Иван любит смотреть   футбол. Мария любит смотреть футбол. Иван и Мария друзья""" # Определяем текст

documents = text.split(".") # Делим текст по точкам

# Создаём объект TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Применяем TF-IDF к текстовым данным
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Получаем списки ключевых слов и их значения TF-IDF для каждого документа
for i in range(len(documents)):
  print(f"Документ {i}: {documents[i]}")
  feature_names = tfidf_vectorizer.get_feature_names_out() # Получаем выделенные слова
  tfidf_scores = tfidf_matrix.toarray()[i] # Получаем массив со значениями TF-IDF
  words_with_scores = [[word, float(score)] for score, word in sorted(zip(tfidf_scores, feature_names), reverse=True)] # Сортируем слова по значимости
  print("Ключевые слова:", words_with_scores)
```
