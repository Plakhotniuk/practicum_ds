```python
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.metrics import pairwise_distances

k_opt = 5  # Оптимальное число кластеров по локтю

pipe = Pipeline(steps=[
    ("prep", preprocess),
    ("kmeans", KMeans(n_clusters=k_opt, random_state=42))
])
pipe.fit(X)  # Обучение K-Means с оптимальным числом кластеров

X_tr = pipe.named_steps["prep"].fit_transform(X)  # Преобразование признаков X для расчёта метрик
labels = pipe.named_steps["kmeans"].labels_
centers = pipe.named_steps["kmeans"].cluster_centers_

wcss = pipe.named_steps["kmeans"].inertia_
sil = silhouette_score(X_tr, labels)

# Межкластерные расстояния между центрами
D = pairwise_distances(centers, metric="euclidean")
mask = ~np.eye(D.shape[0], dtype=bool)  # Убираем диагональ
inter_min = D[mask].min()
inter_mean = D[mask].mean()

print(f"K = {k_opt}")
print(f"WCSS = {wcss:.2f}")
print(f"Silhouette score = {sil:.4f}")
print(f"Межкластерное расстояние: min = {inter_min:.4f}, mean = {inter_mean:.4f}")
```
